```python
import ast

asset_types = set()

for index, row in prompt.iterrows():
    asset_suggestions = ast.literal_eval(row['asset_suggestions'])
    
    # Iterate over asset suggestions for each frame
    for frame_suggestions in asset_suggestions:
        for frame, assets in frame_suggestions.items():
            for asset_type, asset_description in assets.items():
                asset_types.add(asset_type)
                

print("Unique asset types: ")
for asset_type in asset_types:
    print(asset_type)
```

```python
import json

asset_types = set()

# Open and load the JSON file
with open('/home/biniyam/TenAcademy/StoryBoard-Generation/data/concepts.json') as f:
    data = json.load(f)

# Iterate over the rows in the data
for index, row in enumerate(data):
    asset_suggestions = row['asset_suggestions']
    
    # Iterate over asset suggestions for each frame
    for frame_suggestions in asset_suggestions:
        for frame, assets in frame_suggestions.items():
            if isinstance(assets, dict):  # Check if assets is a dictionary
                print(f"Keys for frame {frame}: {list(assets.keys())}")  # Print out the keys
                for asset_type in assets.keys():
                    asset_types.add(asset_type)
            else:
                print(f"Row {index}, frame {frame} could not be parsed")

print("Unique asset types: ")
for asset_type in asset_types:
    print(asset_type)
```
